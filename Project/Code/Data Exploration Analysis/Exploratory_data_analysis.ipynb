### 2. Which company is worth investing based on diluted P/E ratio?

```python
# To calculate the P/E ratio, lets use the diluted EPS which could reflect the actual companies' earning
df_price = df_price.drop(['Unnamed: 0'], axis = 1)
df_price_20161230 = df_price[df_price['date'] == "2016-12-30"]
df_price_20161230 = df_price_20161230.reset_index(drop = True)
df_price_20161230 = df_price_20161230.drop(['open', 'low', 'high', 'volume'], axis = 1)
df_price_20161230.set_index('symbol', inplace = True)

# Create a new table
df_diluted_EPS = df_stats[['Diluted_EPS']]

# Concat both tables together
df_diluted_PE = pd.concat([df_price_20161230, df_diluted_EPS], axis = 1)

# Calculate the diluted P/E ratio
df_diluted_PE = df_diluted_PE.loc[:, 'close'].div(df_diluted_PE['Diluted_EPS'], axis = 0)
print("The company with the lowest Diluted P/E Ratio is ", df_diluted_PE.idxmin(), df_diluted_PE.min().round())
```

### 3. Which company has the highest growth in stock price between 2010 to 2016?


```python
# Take the first date with close price
df_growth = df_price.sort_values('date').groupby('symbol', as_index = False).first()
df_growth2 = df_price.sort_values('date').groupby('symbol', as_index = False).last()
df_growth['Newest_Close'] = df_growth2['close']
df_growth = df_growth.drop(['date', 'open', 'low', 'high', 'volume'], axis = 1)
df_growth['Growth_Percentage'] = (df_growth['Newest_Close'].sub(df_growth['close'], axis = 0)).div(df_growth['close'], axis = 0)*100
df_growth.sort_values('Growth_Percentage', ascending = False).head(1)
```
